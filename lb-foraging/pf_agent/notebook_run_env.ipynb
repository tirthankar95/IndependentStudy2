### Creating neighborhood.

```python
arr = \
[[1, 2, 3, 4, 5, 6], \
 [7, 1, 4, 5, 6, 9], \
 [1, 1, 3, 4, 2, 1], \
 [9, 2, 4, 5, 1, 2], \
 [4, 5, 3, 1, 2, 0], \
 [0, 3, 3, 1, 7, 7]]

x, y = 2, 3
sight = 2
grid_size = 6
for i in range(max(0, x-sight), min(grid_size, x+sight+1)):
    for j in range(max(0, y-sight), min(grid_size, y+sight+1)):
        print(arr[i][j], end = '\t')
    print()
print()
```
